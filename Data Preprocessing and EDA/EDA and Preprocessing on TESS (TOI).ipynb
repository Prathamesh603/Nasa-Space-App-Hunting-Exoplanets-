```python
df_useful.isnull().sum()
```

```python


# Assuming your DataFrame is 'df_useful'

# Separate numeric and object columns
numeric_cols = df_useful.select_dtypes(include=['float64', 'int64']).columns
object_cols = df_useful.select_dtypes(include=['object']).columns

# Fill numeric columns with median
for col in numeric_cols:
    median_val = df_useful[col].median()
    df_useful[col].fillna(median_val, inplace=True)

# Fill object columns with mode
for col in object_cols:
    mode_val = df_useful[col].mode()[0]  # take the most frequent value
    df_useful[col].fillna(mode_val, inplace=True)

# Verify that there are no nulls left
print(df_useful.isnull().sum())
```
